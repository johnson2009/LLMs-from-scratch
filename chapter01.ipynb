绘制简单图形


```python
import numpy as np
import matplotlib.pyplot as plt

#生成数据
x = np.arange(0, 6, 0.1) #以0.1为单位生成0到6的数据
y1 = np.sin(x)
y2 = np.cos(x)

#绘制图形
plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#生成数据
x = np.arange(-6, 6, 0.1) #以0.1为单位生成0到6的数据
y1 = np.exp2(x)
y2 = np.square(x)

#绘制图形
plt.plot(x, y1, label="2^x")
plt.plot(x, y2, label="x^2")

plt.legend()
plt.show()
```

显示4个象限的方法

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(-10, 10, 400)
y = np.sin(x)

# 创建图形
fig, ax = plt.subplots(figsize=(6, 6))

# 绘制曲线
ax.plot(x, y, label='y = sin(x)')

# 设置坐标轴通过原点
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# 隐藏右边和上边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 设置x、y轴刻度
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 设置坐标轴范围
ax.set_xlim(-10, 10)
ax.set_ylim(-2, 2)

# 添加箭头 —— 用 annotate() 模拟
ax.annotate('', xy=(10, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', lw=1.5, color='black'))
ax.annotate('', xy=(0, 2), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', lw=1.5, color='black'))

# 添加坐标轴标签
ax.text(10.2, -0.2, 'x', fontsize=14, color='black')
ax.text(-0.3, 2.1, 'y', fontsize=14, color='black')

# 添加网格（可选）
ax.grid(True, linestyle='--', alpha=0.4)

# 添加图例
ax.legend(loc='upper right')

# 保持比例
ax.set_aspect('equal', adjustable='box')

plt.show()

```
